Trips to Hawaii
-----


```python
from matplotlib import pyplot as plt
```

```python
months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]
revenue = [90,85,70,30,40,40,45,45,40,45,60,80]

plt.figure(figsize=(20,7)) # Only once in beginning

plt.subplot(1,3,1)
plt.plot(months, temperature, marker='*', color='red', linestyle=':')
plt.xlabel('months', fontsize=12)
plt.ylabel('temperature', fontsize=12)
plt.legend('Temprature')
plt.title('Months Vs Temperature', fontsize=14)


plt.subplot(1,3,2)
plt.plot(months, flights_to_hawaii, color='purple', marker='o', linestyle='--')
plt.xlabel('months', fontsize=12)
plt.ylabel('flights_to_hawaii', fontsize=12)
plt.legend('Flights')
plt.title('Months Vs. Flights_to_Hawaii', fontsize=14)


plt.subplot(1,3,3)
plt.plot(months, revenue, color='green', marker='*', linestyle=':')
plt.xlabel('months', fontsize=12)
plt.ylabel('revenue', fontsize=12)
plt.legend('Revenue')
plt.title('Months Vs. Revenue', fontsize=14)

plt.subplots_adjust(left=.125, right=.9, wspace=.5, hspace=.2, top=.9, bottom=.1) # Only once in end

# Default values left=.125, right=.9, wspace=.9, hspace=.2, top=.9, bottom=.1

plt.show()
```
