Sublime Limes' Line Graphs
--
In this project, you will be acting as a data analyst for an online lime retailer called Sublime Limes. People all over the world can use this product to get the freshest, best-quality limes delivered to their door. One of the product managers at Sublime Limes would like to gain insight into the customers and their sales habits. Using Matplotlib, you'll create some different line graphs (or maybe, lime graphs) to communicate this information effectively.

```python
from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# lime species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]

# Strawbery species sold each month
key_strawberies_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
persian_strawberies_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_strawberies_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]

x_values = range(len(months))

plt.figure(figsize=(30,15))


# Key Limes
ax = plt.subplot(2,3,1)
ax.set_xticks(x_values)
ax.set_xticklabels(months)
plt.plot(x_values, key_limes_per_month, color='blue', marker='*', markersize=10, linestyle='--', linewidth=3)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.title('Total Sales KEY LIMES', fontsize=12)
legend_label = ["Key Lime"] 
plt.legend(legend_label)


# Persian Limes
ax = plt.subplot(2,3,2)
ax.set_xticks(x_values)
ax.set_xticklabels(months)
x_values = range(len(months))
plt.plot(x_values, persian_limes_per_month, color='purple', marker='s', markersize=10, linestyle=':', linewidth=3)
plt.xlabel('Months',fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.title('Total sales PERSIAN LIMES', fontsize=12)
legend_label = ["Persian Lime"] 
plt.legend(legend_label)


# Blood Limes
ax = plt.subplot(2,3,3)
ax.set_xticks(x_values)
ax.set_xticklabels(months)
plt.plot(x_values, blood_limes_per_month, color='red', marker='*', markersize=10, linestyle=':', linewidth=3)
plt.xlabel('Months',fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.title('Total sales BLOOD LIMES', fontsize=12)
legend_label = ["Blood Lime"] 
plt.legend(legend_label)


# Key Strawberies
ax = plt.subplot(2,3,4)
ax.set_xticks(x_values)
ax.set_xticklabels(months)
plt.plot(x_values, key_strawberies_per_month, color='green', marker='o', markersize=10, linestyle='--', linewidth=3)
plt.xlabel('Months',fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.title('Total sales for KEY STRAWBERIES', fontsize=12)
legend_label = ["Key Strawberies"] 
plt.legend(legend_label)


# Persian Strawberies
ax = plt.subplot(2,3,5)
ax.set_xticks(x_values)
ax.set_xticklabels(months)
plt.plot(x_values, persian_strawberies_per_month, color='black', marker='s', markersize=10, linestyle=':', linewidth=3)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.title('Total sales for PERSIAN STRAWBERIES', fontsize=12)
legend_label = ["Persian Strawberies"] 
plt.legend(legend_label)


# Blood Strawberies
ax = plt.subplot(2,3,6)
ax.set_xticks(x_values)
ax.set_xticklabels(months)
plt.plot(x_values, blood_strawberies_per_month, color='red', marker='*', markersize=10, linestyle=':', linewidth=3)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.title('Total sales for BLOOD STRAWBERIES', fontsize=12)
legend_label = ["Blood Strawberies"] 
plt.legend(legend_label)

plt.subplots_adjust(wspace=0.5)
plt.savefig('Sublime_Sales.png')
plt.show()

```
